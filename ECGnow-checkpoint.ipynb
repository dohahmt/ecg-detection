```python
import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
from keras import layers
from keras import models
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python
dataset = pd.read_csv("C:/Users/doha_/Downloads/ecg - Copy.csv", header=None)
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.values # prendre le dataset sans les entêtes
# divisé le dataset en train 80% et test 20%
train_X, test_X = train_test_split(
    X, test_size=0.2, random_state=22
)
print(train_X.shape)
print(test_X.shape)
```

```python
# normalisation des données entre 0 et 1
min_val = tf.reduce_min(train_X)
max_val = tf.reduce_max(train_X)

train_X = (train_X - min_val) / (max_val - min_val)
test_X = (test_X - min_val) / (max_val - min_val)

train_X = tf.cast(train_X, tf.float32)
test_X = tf.cast(test_X, tf.float32)
```

```python
# visualisation de la premiere observation (le premier battement de coeur)
plt.grid()
plt.plot(np.arange(140), train_X[0])
plt.title("ECG")
plt.show()
```

```python
# le modèle: autoencodeur 140(input shape) -> 32 -> 32/2 -> 32/4(bottleneck) -> 32/2 -> 32 -> 140

input_layer = layers.Input(shape=(train_X.shape[1],))
#Encoder
encoder = layers.Dense(32, activation="tanh")(input_layer)
encoder = layers.Dense(16, activation='relu')(encoder)
encoder = layers.Dense(8, activation='relu')(encoder)

encoderModel = tf.keras.Model(input_layer, encoder, name="encoder")
#Decoder
decoder = layers.Dense(16, activation='relu')(encoder)
decoder = layers.Dense(32, activation='relu')(decoder)
decoder = layers.Dense(train_X.shape[1], activation='tanh')(decoder)

#decoderModel = Model(decoder_input, decoder, name="decoder")
#Autoencoder
autoencoder = tf.keras.Model(inputs=input_layer, outputs=decoder, name="autoencoder")
autoencoder.compile(optimizer="adam", loss="mse") 
autoencoder.summary()
```

```python
history = autoencoder.fit(train_X, train_X, 
          epochs=20, 
          batch_size=512,
          validation_data=(test_X, test_X),
          shuffle=True)
```

```python
plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.title('Model loss')
plt.legend()
```

```python
reconstructions = autoencoder.predict(train_X)

plt.plot(train_X[0], 'b')
plt.plot(reconstructions[0], 'r')
plt.fill_between(np.arange(140), reconstructions[0], train_X[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.title('Erreur de reconstruction')
plt.show()
```

```python
test_X_predictions = autoencoder.predict(test_X)

plt.plot(test_X[0], 'b')
plt.plot(test_X_predictions[0], 'r')
plt.fill_between(np.arange(140), test_X_predictions[0], test_X[0], color='lightcoral')
plt.legend(labels=["Input", "Reconstruction", "Error"])
plt.show()
```

```python
trainMSE = np.mean(np.power(train_X - reconstructions, 2), axis=1)
plt.hist(trainMSE, bins=80)
plt.title('Histogramme de l\'erreur de reconstruction')
```

```python
threshold = np.mean(trainMSE) + (np.std(trainMSE))
print("Threshold: ", threshold)
```

```python
# visuelement le seuil est:
threshold=0.0043
```

```python
decision_df = pd.DataFrame()
decision_df['trainMSE'] = trainMSE
decision_df['threshold'] = threshold
decision_df['anomaly'] = decision_df['trainMSE'] > decision_df['threshold']
#display(decision_df)
#pd.set_option("display.max_rows", None, "display.max_columns", None)

print(decision_df)
```

```python
groups = decision_df.groupby('anomaly')
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.index, group.trainMSE, marker='o', ms=3.5, linestyle='',
            label= "Anomalie" if name == 1 else "Normal")
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Erreur de reconstruction et seuil des décision")
plt.ylabel("Erreur de reconstruction")
plt.xlabel("Indice des points de données")
plt.show();
```

```python
testMSE = np.mean(np.power(test_X - test_X_predictions, 2), axis=1)
plt.hist(testMSE, bins=80)
plt.title('Histogramme de l\'erreur de reconstruction')
```

```python
decision_df = pd.DataFrame()
decision_df['testMSE'] = testMSE
decision_df['threshold'] = threshold
decision_df['anomaly'] = decision_df['testMSE'] > decision_df['threshold']
#display(decision_df)
pd.set_option("display.max_rows", None, "display.max_columns", None)

print(decision_df)
```

```python
groups = decision_df.groupby('anomaly')
fig, ax = plt.subplots()
for name, group in groups:
    ax.plot(group.index, group.testMSE, marker='o', ms=3.5, linestyle='',
            label= "Anomaly" if name == 1 else "Normal")
ax.hlines(threshold, ax.get_xlim()[0], ax.get_xlim()[1], colors="r", zorder=100, label='Threshold')
ax.legend()
plt.title("Erreur de reconstruction et seuil de décision")
plt.ylabel("Erreur de reconstruction")
plt.xlabel("Indice des points de données")
plt.show();
```
